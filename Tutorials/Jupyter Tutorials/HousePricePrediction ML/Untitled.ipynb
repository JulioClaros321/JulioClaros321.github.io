```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set() 
%matplotlib inline
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
all_df = pd.concat([train.drop("SalePrice", axis=1), test], axis=0, sort=False)
all_df.head() 
```

```python
len(all_df)
```

```python
nan_count = all_df.isnull().sum().sort_values(ascending=False)[all_df.isnull().sum() > 0]
nan_count.head()
```

```python
data_type = all_df.dtypes
data_type.head()
```

```python
nan_table = pd.concat([nan_count, data_type],
                     axis=1,
                     keys=["nan_count", "data_type"])
nan_table
```

```python
nan_table.dropna(inplace=True)
```

```python
nan_table = nan_table.sort_values(["data_type", "nan_count"])
nan_table
```

```python
nan_table[["favourable_value", "nan_percentage"]]= ["", 0.0]
nan_table
```

```python
for i, row in nan_table.iterrows():
    #finds percentage of nan_count
    nan_table.at[i, "nan_percentage"] = (nan_table.loc[i].nan_count/len(all_df)) * 100
    
    if all_df[i].dtype == "float64": 
        nan_table.at[i, "favourable_value"] = all_df[i].median() 
    
    elif all_df[i].dtype == "object": 
        nan_table.at[i, "favourable_value"] = all_df[i].mode().values[0]

nan_table
```

```python
arr_na = ["BsmtFinType1", "BsmtFinType2", "BsmtQual", "BsmtExposure", 
          "BsmtCond", "GarageType", "GarageFinish", "GarageQul",
         "GarageCond", "FireplaceQu", "Fence", "Alley", "MiscFeature", "PoolQC"]

def search(array, data): 
    for i in range(len(array)): 
        if array[i] == data:
            return True
        return False 

for i, row in nan_table.iterrows():
    if nan_table.loc[i].data_type == "float64":
        
        all_df[i].fillna(nan_table.loc[i].favourable_value, inplace=True) 
        
    else: 
        if search(arr_na, i):
            all_df[i].fillna("NA", inplace=True)
                
            nan_table.at[i, "favourable_value"] = "NA"
        
        else: 
            all_df[i].fillna(nan_table.loc[i].favourable_value, inplace=True)
            
        
```
