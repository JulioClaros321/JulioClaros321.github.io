```python
import pandas 
import pyautogui
import time
from datetime import date
```

```python
#pyautogui.moveTo(800, 15) # Word Tab
#pyautogui.moveTo(800, 500) # Touch word 

#pyautogui.moveTo(500, 15) # Click Indeed Tab
#pyautogui.moveTo(1075, 325, .5) # Click Job Div
#pyautogui.moveTo(500, 900) # Click First Job On Page
#pyautogui.moveT1o(600, 300) # Click Empty Part Of Page To Reset Alt

#pyautogui.moveTo(80, 265) # Click New Document One Drive Tab * 3 for word
#pyautogui.moveTo(1020, 180) # Elements Tab On Inspect


def generate_page_title():
    report_date = date.today()
    job_title = input('Job Title: ')
    state = input('State Of Search: ')
    total_jobs_found = int(input('Numb. Of Total Jobs Found: '))
    
    total_pages = total_jobs_found // 15
    remainder = total_jobs_found % 15 
    
    title = str(job_title).lower() + '_' + str(state).lower() + '_' + str(report_date)
    
    return (title, total_pages, remainder)

def open_new_word_doc(title):
    time.sleep(4)
    pyautogui.moveTo(800, 15) # Word Tab
    time.sleep(2)
    pyautogui.click()

    with pyautogui.hold('alt'):
        time.sleep(2)
        pyautogui.press('d')
        time.sleep(2)

    pyautogui.typewrite(r'https://eseal-my.sharepoint.com/personal/jclaros_eseal_org/')
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(7)

    pyautogui.moveTo(80, 265) # Click New Document One Drive Tab * 3 for word
    time.sleep(2)
    pyautogui.click()
    time.sleep(2)

    pyautogui.press(['tab'] * 3) 
    time.sleep(2)
    pyautogui.press('enter')
    time.sleep(5)
    
    with pyautogui.hold('ctrl'):
        time.sleep(1)
        with pyautogui.hold('shift'):
            time.sleep(2)
            pyautogui.press('tab')
    
    with pyautogui.hold('ctrl'):
        time.sleep(2)
        pyautogui.press('w')
        time.sleep(2)
    

    pyautogui.moveTo(800, 500, 1) # Touch word 
    time.sleep(1)
    pyautogui.click()
    
    time.sleep(3)
    pyautogui.typewrite(str(title))
    time.sleep(2)
    pyautogui.press(['enter'] * 3) 
    
    time.sleep(10)

def job_div_paste_function(counter, total_jobs_done):
    if counter == 0:
        time.sleep(3)
        pyautogui.moveTo(500, 15) # Click Indeed Tab
        time.sleep(1)
        pyautogui.click() 
        time.sleep(2)
        pyautogui.moveTo(500, 900) # Click First Job On Page
        time.sleep(2)
        pyautogui.click() 
        time.sleep(10)
    else: 
        time.sleep(3)
        pyautogui.press(['tab'] * 2)
        time.sleep(2)
        pyautogui.press('enter')
        time.sleep(10)
    

    pyautogui.moveTo(1020, 180) # Elements Tab On Inspect
    time.sleep(2)
    pyautogui.click()
    pyautogui.moveTo(1075, 325, .5) # Click Job Div
    time.sleep(3)
    pyautogui.click()
    time.sleep(3)

    with pyautogui.hold('ctrl'):
        time.sleep(2)
        pyautogui.press('c')
        
    time.sleep(3)


    pyautogui.moveTo(800, 15) # Word Tab
    time.sleep(2)
    pyautogui.click()
    time.sleep(2)

    with pyautogui.hold('ctrl'):
        time.sleep(2)
        pyautogui.press('v')
    
    time.sleep(6)

    pyautogui.press(['enter'] * 2) 
    time.sleep(3)
    pyautogui.typewrite('--------------------' + ' Total Jobs Done: ' + str(total_jobs_done))
    time.sleep(3)
    pyautogui.press(['enter'] * 2) 

    time.sleep(2)

    pyautogui.moveTo(500, 15) # Click Indeed Tab
    time.sleep(2)
    pyautogui.click()
    time.sleep(2)
    pyautogui.moveTo(600, 300) # Click Empty Part Of Page To Reset Alt
    time.sleep(3)
    pyautogui.click()
    time.sleep(1)

    with pyautogui.hold('alt'):
        time.sleep(1)
        pyautogui.press('left')
        
    time.sleep(10)

def main_loop(title, page_numb, remainder):
    jobs_per_page = 15
    pages_per_word_doc = 10
    total_indeed_pages_completed = 0
    word_page_counter = 1
    iterations = 0
    total_jobs_done = 0
    
    while total_indeed_pages_completed < int(page_numb)- 3:
        total_pages_pasted = 0
        
        open_new_word_doc(title + '_page' + str(word_page_counter))
        
        while total_pages_pasted < 5:
            counter = 0
            print(counter)
            while counter < jobs_per_page:
                if iterations > 0: new_page_icon = 8
                else: new_page_icon = 7
                    
                job_div_paste_function(counter, total_jobs_done)
                
                if counter == 14:
                    
                    pyautogui.press(['tab'] * new_page_icon)
                    time.sleep(2)
                    pyautogui.press('enter')
                    time.sleep(10)
                    
                total_jobs_done += 1  
                counter += 1
                
            iterations += 1  
            total_pages_pasted += 1
            total_indeed_pages_completed += 1
        
        word_page_counter += 1   
```

```python
def main(): 
    variables = generate_page_title()
    main_loop(variables[0], variables[1], variables[2])
```

```python
main()
```
